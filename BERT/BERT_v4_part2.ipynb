```python
import pandas as pd
```

```python
df = pd.read_csv("BERTv4_result.csv", index_col=0)
```

```python
df.columns = ["toxic", "severe_toxic", "obscene", "threat", "insult", "identity_hate"]
df
```

```python
df['total_score'] = df.sum(axis=1)
```

```python
df
```

```python
less_toxic_score = []
for i in range(int(len(df)/2)):
    less_toxic_score.append(df.total_score[i])
```

```python
len(less_toxic_score)
```

```python
more_toxic_score = []
for i in range(int(len(df)/2)):
    more_toxic_score.append(df.total_score[i+30108])
```

```python
len(more_toxic_score)
```

```python
validation = pd.read_csv("../dataset/validation_data.csv")
```

```python
validation
```

```python
validation["less_toxic_score"] = less_toxic_score
validation["more_toxic_score"] = more_toxic_score
```

```python
validation
```

```python
def function(x):
    if x.less_toxic_score < x.more_toxic_score:
        return 1
    else:
        return 0
    
validation["comparison"] = validation[["less_toxic_score", "more_toxic_score"]].apply(function, axis=1)
```
